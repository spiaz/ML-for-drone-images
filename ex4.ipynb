The corruptedImage.png image is missing 87.5% of its pixels (they have been set to 0). 

Can you recover their RGB values? 

Please describe your approach, provide your reconstructed image and the code to reproduce your result.

## Execution

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [20, 15]

img = cv2.imread("data/corruptedImage.png")

_ = plt.imshow(img)
```

```python
kernel = np.ones((5, 5), np.uint8)
dil = cv2.dilate(img, kernel, iterations=1)
kernel = np.ones((2, 2), np.uint8)
dil = cv2.dilate(dil, kernel, iterations=2)

med = cv2.medianBlur(dil, 5)

_ = plt.imshow(med)
```
